```python
# 10 pages data collecting down

final=pd.DataFrame()

for j in range(1,11):
    
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    webpage = requests.get(url).text
    
    soup= BeautifulSoup(webpage)
    
    company=soup.find_all('div',class_='company-content-wrapper')
    
    company_name=[]
    rating=[]
    reviews=[]
    sector=[]
    hq=[]
    old=[]
    employees=[]

    for i in company :
        try:
            company_name.append(i.find('h2').text.strip())
        except:
            company.append(np.nan)
            
        try:
            rating.append(i.find('p',class_='rating').text.strip())
        except:
            rating.append(np.nan)
            
        try:
            reviews.append(i.find_all('a', class_= 'review-count')[0].text.strip())
        except:
            reviews.append(np.nan)
        
        try:
            sector.append(i.find_all('p',class_='infoEntity sbold-list-header')[0].text.strip())
        except:
            sector.appen(np.nan)
            
        try:
            hq.append(i.find_all('p',class_='infoEntity sbold-list-header')[1].text.strip())
        except:
            hq.append(np.nan)
        
        try:
            old.append(i.find_all('p',class_='infoEntity sbold-list-header')[2].text.strip())
        except:
            old.append(np.nan)
        
        try:
            employees.append(i.find_all('p',class_='infoEntity sbold-list-header')[3].text.strip())
        except:
            employees.append(np.nan)
        
    d={
        'name':company_name,
        'rating':rating,
        'reviews':reviews,
        'sector':sector,
        'hq':hq,
        'old':old,
        'employees':employees
    }

    df = pd.DataFrame(d)
    
    final = final.append(df)
```
